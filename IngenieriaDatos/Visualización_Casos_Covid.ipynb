```python
# Cambio de formato de fechas
datos.Fecha_de_diagnóstico=pd.to_datetime(datos.Fecha_de_diagnóstico,dayfirst='True').dt.strftime('%Y%m%d')
datos.fecha_reporte_web=pd.to_datetime(datos.fecha_reporte_web,dayfirst='True').dt.strftime('%Y%m%d')
datos.Fecha_de_inicio_de_síntomas=pd.to_datetime(datos.Fecha_de_inicio_de_síntomas,dayfirst='True').dt.strftime('%Y%m%d')
datos.Fecha_de_muerte=pd.to_datetime(datos.Fecha_de_muerte,dayfirst='True').dt.strftime('%Y%m%d')
datos.Fecha_de_notificación=pd.to_datetime(datos.Fecha_de_notificación,dayfirst='True').dt.strftime('%Y%m%d')
datos.Fecha_de_recuperación=pd.to_datetime(datos.Fecha_de_recuperación,dayfirst='True').dt.strftime('%Y%m%d')
```

```python
# Cambio de tipo de dato a las fechas "datetime64"
datos.Fecha_de_diagnóstico=datos.Fecha_de_diagnóstico.astype("datetime64")
datos.fecha_reporte_web=datos.fecha_reporte_web.astype("datetime64")
datos.Fecha_de_inicio_de_síntomas=datos.Fecha_de_inicio_de_síntomas.astype("datetime64")
datos.Fecha_de_muerte=datos.Fecha_de_muerte.astype("datetime64")
datos.Fecha_de_notificación=datos.Fecha_de_notificación.astype("datetime64")
datos.Fecha_de_recuperación=datos.Fecha_de_recuperación.astype("datetime64")
```

```python
# Fecha de referencia: Fecha de inicio de síntomas del primer caso
datos.Fecha_de_diagnóstico=datos.Fecha_de_diagnóstico-min(datos.Fecha_de_inicio_de_síntomas)
datos.fecha_reporte_web=datos.fecha_reporte_web-min(datos.Fecha_de_inicio_de_síntomas)
datos.Fecha_de_notificación=datos.Fecha_de_notificación-min(datos.Fecha_de_inicio_de_síntomas)
datos.Fecha_de_muerte=datos.Fecha_de_muerte-min(datos.Fecha_de_inicio_de_síntomas)
datos.Fecha_de_recuperación=datos.Fecha_de_recuperación-min(datos.Fecha_de_inicio_de_síntomas)
# Dejar esta transformación de última, dado que es la fecha de referencia
datos.Fecha_de_inicio_de_síntomas=datos.Fecha_de_inicio_de_síntomas-min(datos.Fecha_de_inicio_de_síntomas)
```

```python
# Pasar a escalar (o enteros) las variables 'fechas'
datos.Fecha_de_diagnóstico=datos.Fecha_de_diagnóstico.dt.days
datos.fecha_reporte_web=datos.fecha_reporte_web.dt.days
datos.Fecha_de_notificación=datos.Fecha_de_notificación.dt.days
datos.Fecha_de_muerte=datos.Fecha_de_muerte.dt.days
datos.Fecha_de_recuperación=datos.Fecha_de_recuperación.dt.days
datos.Fecha_de_inicio_de_síntomas=datos.Fecha_de_inicio_de_síntomas.dt.days
```

```python
# 1. Graficar cómo ha evolucionado el Covid-19 en Colombia
# Necesitamos un histograma de frecuencias por fecha. Los histogramas necesitan un argumento que es el número de "BINS"
# o barras, la estadística dice que barras o bins= raíz(n)
barras=int(datos.Fecha_de_inicio_de_síntomas.max())
barras
```

```python
plt.rcParams['figure.figsize']=(7,5) # Pulgadas
plt.hist(datos.Fecha_de_inicio_de_síntomas,bins=barras)
print("Evolución del Covid-19 en Colombia")
```

```python
# Cómo saber si se llegó a una meseta "Inmunudad de rebaño"---> Graficamos el acumulado
plt.rcParams['figure.figsize']=(7,5)
plt.hist(datos.Fecha_de_inicio_de_síntomas,bins=barras,cumulative=True)
print("Evolución del Covid-19 en Colombia")
```

```python
# 5. Hacer un zoom, departamentos y municipios
# Antioquia y Bogotá (contagiados)
datos_Antioquia=datos[(datos.Nombre_departamento=="ANTIOQUIA")]
datos_Bogotá=datos[(datos.Nombre_departamento=="BOGOTA")]
```
